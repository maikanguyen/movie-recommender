```python
from pyspark import SparkConf, SparkContext
import pandas as pd
```

```python
conf = SparkConf().setMaster("local[*]").setAppName("MovieRecommender")
sc = SparkContext(conf=conf)
```

```python
data = sc.textFile("data/ml-100k/u.data")
```

```python
data.collect()
```

```python
movie_users = data.flatMap(lambda line: line.split())
```

```python
movie_tuple = movie_users.map(lambda movie: (movie, 1))
movie_tuple
```

```python
results_count= movie_tuple.reduceByKey(lambda v1,v2: (v1+v2))
```

```python
mSorted = results_count.sortBy(lambda a: -a[1])
mSorted.collect()

mTop10 = mSorted.top(10, key=lambda x: x[1])
mTop10
```

```python
data1 = sc.textFile("data/ml-100k/u.item")
```

```python
data1.collect()
```

```python
movies = data1.map(lambda line: (line.split('|')[0], line.split('|')[1]))
movies.takeSample(False, 5)
movies_dict = movies.collectAsMap()
```

```python
movies_dict
```

```python
print([ (movies_dict.get(r[0]),r[1]) for r in mTop10 ])
```
